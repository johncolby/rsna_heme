# save_frames

Save some example images for visual inspection

```python
import mxnet as mx
import os

from matplotlib import pyplot as plt
from mxnet.gluon.data.vision import datasets
from tqdm import tqdm_notebook as tqdm

import rsna_heme as rsna
```

```python
base_dir = '/mnt/Data2/datasets/rsna_heme/normalized'
out_dir  = 'frames'
```

```python
train_dataset = datasets.ImageRecordDataset(os.path.join(base_dir, 'train.rec'), flag=1, transform=rsna.transforms.common_transform)
```

```python
plt.figure(figsize=(5.12,5.12), dpi=100)
plt.imshow(train_dataset[0][0].asnumpy())
```

```python
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

for i in tqdm(range(1000)):
    out_path = os.path.join(out_dir, f'{i:03d}.png')
    plt.imsave(out_path, train_dataset[i][0].asnumpy())
```
